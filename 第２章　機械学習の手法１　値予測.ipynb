```python
import pandas as pd
#身長と体重のデータを読み込む
children = pd.read_csv("https://raw.githubusercontent.com/futurebridge/ZukaiML/master/children.csv",encoding="SHIFT-JIS")

#読み込んだデータの表示
children
```

```python
import numpy as np

#x,yの平均を求める
x,y = np.mean(children)
print (x,y)

```

```python
#共分散を求める
#データ個数分だけ (x-x平均)x(y-y平均)を繰り返す
sum=0
for i in range(0,len(children)):
  sum += ((children["年齢"][i]-x)*(children["身長"][i]-y))
  print(i,"年齢",children["年齢"][i],"身長",children["身長"][i],sum)

#データ個数で割った値を返す
cov_a = sum/len(children)

print ("共分散:",cov_a)
```

```python

#年齢と身長の標準偏差を求める
a_std,b_std=np.std(children)

#相関係数を求める
#相関係数 = (共分散)/(身長の標準偏差x年齢の標準偏差)

print ("相関係数:",cov_a/(a_std*b_std))

```

```python
#最小二乗法による傾きa と　切片bを求める

#分散を求める
a_var,b_var = np.var(children)

#傾きa = 共分散/分散
a = cov_a/a_var

#切片 b= y平均 - a x x平均
b = y - a * x

print ("傾きa",a, "切片b",b)
```

```python
import pandas as pd
#身長と体重のデータを読み込む
children = pd.read_csv("https://raw.githubusercontent.com/futurebridge/ZukaiML/master/children.csv",encoding="SHIFT-JIS")

#読み込んだデータの表示
children
```

```python

#説明変数の抽出
X = children["年齢"] 
#目的変数の抽出
Y = children["身長"] 


#ライブラリの導入
#エラーになる場合、anaconda promptからpip install statsmodels
import statsmodels.api as smf
#定数表示
X1=smf.add_constant(X) 
#OLS(最小二乗法)でモデル作成
model = smf.OLS(Y,X1)
#model.fitで実際の計算
result = model.fit()
#結果表示
result.summary()
```

```python
##決定係数
#残差平方和,回帰平方和、全体平方和を求める
z_sum=0
k_sum=0
t_sum=0
for i in range(0,len(children)):
    #残差平方和 y(実測値) - 予測値
    z_sum += (children["身長"][i]-(children["年齢"][i]*a+b))**2
    #回帰平方和 予測値 - 平均
    k_sum+= ((children["年齢"][i]*a+b)-y)**2
    #全体平方和 y(実測値) - 平均
    t_sum+= ((children["身長"][i])-y)**2


    print ("残差平方和:",z_sum,"回帰平方和:",k_sum,"全体平方和:",t_sum)

#決定係数を求める
print ("決定係数", k_sum/t_sum, 1 - z_sum/t_sum)


```

```python
#標準誤差を求める

#残差平方和を自由度を引いた個数でかける

se1=z_sum*(1/(len(children)-2))

# x-x平均の平方和
x_sum=0
for i in range(0,len(children)):
    x_sum += (children["年齢"][i]-x)**2

se2=np.sqrt((se1/x_sum))
    
#標準誤差を求める
print ("標準誤差:",se2)

```

```python
#t値を求める 
tval = a/se2
print ("t値'",tval)
```
